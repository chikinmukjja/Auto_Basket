```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_excel("basket_data_roi.xlsx")
train.head(3)
```

```python
from sklearn.cross_validation import cross_val_score

feature_names = train.columns.drop('lable')
label_name = ['lable']

train_X = train[feature_names]
train_y = train[label_name]
```

```python
from sklearn.svm import LinearSVC
from sklearn.grid_search import GridSearchCV
from sklearn.svm import SVC
```

```python
svc_param = {'C':np.logspace(-2,0,20)}
svc_rbf_param = [{'C':np.logspace(-2,0,15),"gamma":np.logspace(-2,0,15),'kernel': ['rbf']}]
svc_poly_param = [{'C':np.logspace(-2,0,10), 'gamma': np.logspace(-2,0,10), 'kernel': ['poly'], 'degree': [2,3,4,5]}]
```

### find hypterparameter

```python
gs_svc = GridSearchCV(LinearSVC(loss='l2'),svc_param,cv=5,n_jobs=4,verbose=1)
```

```python
gs_svc.fit(np.array(train_X),np.array(train_y['lable']))
```

```python
gs_svc_poly = GridSearchCV(SVC(), svc_poly_param,cv=5,n_jobs=4,verbose=1)
```

```python
gs_svc_poly.fit(np.array(train_X),np.array(train_y['lable']))
```

```python
gs_svc_rbf = GridSearchCV(SVC(), svc_rbf_param,cv=5,n_jobs=4,verbose=1)
```

```python
gs_svc_rbf.fit(np.array(train_X),np.array(train_y['lable']))
```
